```python
import scipy.stats as ss
import matplotlib.pyplot as plt
import numpy as np

ticks = 720
beta_rvs = [np.int(ss.beta(1, 3).rvs() * ticks) for i in range(200)]
plt.hist(np.asarray(beta_rvs))
```
