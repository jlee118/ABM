```python
data_dir = "C:\\Users\\jchle\\PhD\\ABM\\results2\\14-12-2022_02_17_09_docs_6_nurses_3_alt_0"
dataset = pd.read_csv(os.path.join(data_dir, "stats.csv"))
data_dir2 = "C:\\Users\\jchle\\PhD\\ABM\\results2\\14-12-2022_02_17_14_docs_6_nurses_3_alt_3"
dataset2 = pd.read_csv(os.path.join(data_dir2, "stats.csv"))
data_dir3 = "C:\\Users\\jchle\\PhD\\ABM\\results2\\14-12-2022_02_17_21_docs_6_nurses_3_alt_5"
dataset3 = pd.read_csv(os.path.join(data_dir3, "stats.csv"))

d1 = dataset[['Low Priority Queue Size', 'High Priority Queue Size']].rename(columns={'Low Priority Queue Size': 'Low Priority Queue Size 0', 'High Priority Queue Size': 'High Priority Queue Size 0'})
d2 = dataset2[['Low Priority Queue Size', 'High Priority Queue Size']].rename(columns={'Low Priority Queue Size':'Low Priority Queue Size 3', 'High Priority Queue Size': 'High Priority Queue Size 3'})
d3 = dataset3[['Low Priority Queue Size', 'High Priority Queue Size']].rename(columns={'Low Priority Queue Size':'Low Priority Queue Size 5', 'High Priority Queue Size': 'High Priority Queue Size 5'})
final = pd.concat([d1, d2, d3], axis=1)
```
